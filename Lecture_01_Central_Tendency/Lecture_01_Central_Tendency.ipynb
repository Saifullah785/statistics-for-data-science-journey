```python
# Import necessary libraries for numerical operations and plotting
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Generate sample data and calculate quartiles and IQR
np.random.seed(0)
data = np.random.normal(100, 20, 200)
data = np.append(data, [300, 5])

Q1 = np.percentile(data, 25)
Q3 = np.percentile(data, 75)
IQR = Q3 - Q1
```

```python
# Define bounds and remove outliers from the data
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identifying outliers
outliers = (data < lower_bound) | (data > upper_bound)

# Removing outliers
data_no_outliers = data[~outliers]
```

```python
# Visualize original data and data without outliers using box plots
plt.figure(figsize=(14, 6))
plt.subplot(1, 2, 1)
sns.boxplot(data)
plt.title('Original Data with Outliers')

plt.subplot(1, 2, 2)
sns.boxplot(data = data_no_outliers)
plt.title('Data without Outliers')

plt.tight_layout()
```

```python
# Calculate the variance of the given array
import numpy as np

df = np.array([2,3,4,5,6])
mean = np.mean(df)
var = np.var(df, ddof=1)
var
```

```python
# Calculate the standard deviation and standard error of the array
import numpy as np

df = np.array([2,3,4,5,6])

std = np.std(df, ddof=1)
std

print("Standard Deviation: ", std)

se = std / np.sqrt(len(df))
print("Standard Error: ", se)
```
