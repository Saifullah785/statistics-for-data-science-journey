**Chi-Squared Test**

```python
# Perform a one-sample Chi-Squared test
import numpy as np
from scipy.stats import chisquare

observed_frequencies = np.array([30, 70])

total_meals = observed_frequencies.sum()
expected_frequencies = np.array([total_meals * 0.5, total_meals * 0.5])

chi2_stat, p_value = chisquare(observed_frequencies, f_exp=expected_frequencies)

print(f"Chi-Squared Statistic: {chi2_stat}")
print(f"P-Value: {p_value}")

if p_value < 0.05:
    print("Reject the null hypothesis. There is a significant difference between the groups.")
else:
    print("Fail to reject the null hypothesis, there is no significant difference between the groups.")
```

```python
# Load the 'tips' dataset and display its head
import pandas as pd
import numpy as np
import seaborn as sns
from scipy.stats import chi2_contingency

df = sns.load_dataset('tips')
df.head()
```

```python
# Create a contingency table for 'sex' and 'day'
contigency_table = pd.crosstab(df['sex'], df['day'])
contigency_table
```

```python
# Perform Chi-Squared test on the contingency table
chi2, p, dof, expected = chi2_contingency(contigency_table)

print(f"Chi-Squared Statistic: {chi2}")
print(f"P-Value: {p}")
print(f"Degrees of Freedom: {dof}")
print("Expected: \n", np.array(expected))


if p < 0.05:
  print("Reject the null hypothesis. There is a significant difference between the groups.")
else:
  print("Fail to reject the null hypothesis, there is no significant difference between the groups.")
```


# **t-test**

**1.one sample t-test**

```python
# Import libraries for t-test analysis
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
```

```python
# Load the 'titanic' dataset and display its head
df = sns.load_dataset('titanic')
df.head()
```

```python
# Plot histogram of age with KDE
sns.histplot(df['age'], kde=True)
```

```python
# Fill missing 'age' and 'fare' values with their medians
df['age'] = df['age'].fillna(df['age'].median())
df['fare'] = df['fare'].fillna(df['fare'].median())
```

```python
# Perform Shapiro-Wilk test for normality on 'age'
statistic, p = stats.shapiro(df['age'])

if p < 0.05:
  print("Reject the null hypothesis. The data is not normally distributed.")
else:
  print("Fail to reject the null hypothesis. The data is normally distributed")
```

```python
# Calculate the mean of the 'age' column
df['age'].mean()
```

```python
# Perform a one-sample t-test on 'age' against a population mean of 35
statistic, p = stats.ttest_1samp(df['age'], 35)

if p < 0.05:
  print("Reject the null hypothesis. There is a significant difference between the groups.")
else:
  print("Fail to reject the null hypothesis, there is no significant difference between the groups.")
```

# **2. Two sample independent t-test**

```python
# Perform an independent two-sample t-test on age between male and female passengers
age_male = df[df['sex'] == 'male']['age']
age_female = df[df['sex'] == 'female']['age']

statistic, p = stats.ttest_ind(age_male, age_female)

if p < 0.05:
    print(f"Reject null hypothesis, because p vaule: {p} , there is a significant difference between ages of male and female passengers of titanic.")
else:
    print(f"Fail to reject null hypothesis,because p vaule: {p}, there is not any differences between ages of female and male passengers.")
```

# **one-way ANOVA**

1. There is a difference between ages of First, second and third class passengers?

H0: There is no difference.

H1: There is a significant Difference.

```python
# Display 'age' and 'pclass' columns
df[['age', 'pclass']].head()
```

```python
# Perform one-way ANOVA to compare age across passenger classes
import statsmodels.api as sm
from statsmodels.formula.api import ols

model = ols('age ~ C(pclass)', data=df).fit()
anova_table = sm.stats.anova_lm(model, typ=2)
print(anova_table)

if anova_table['PR(>F)'][0] < 0.05:
  print("Reject null hypothesis. because p value: {anova_table['PR(>F)'][0]} , there is a significant difference between ages of passengers in different classes.")
else:
  print("Fail to reject null hypothesis, because p value: {anova_table['PR(>F)'][0]}, there is no significant difference between ages of passengers in different classes.")
```

# **PostHoc Test**

```python
# Perform Tukey's HSD post-hoc test for pairwise comparisons of age across pclass groups
from statsmodels.stats.multicomp import pairwise_tukeyhsd

tukey = pairwise_tukeyhsd(endog=df['age'], groups=df['pclass'], alpha=0.05)
tukey.summary()
```

# **TWO Way ANOVA**

```python
# Display the head of the DataFrame
df.head()
```

```python
# Perform two-way ANOVA to examine the effects of sex and who on age
import statsmodels.api as sm
from statsmodels.formula.api import ols

model = ols("age ~ sex * who", data=df).fit()
anova_table = sm.stats.anova_lm(model, typ=2)
print(anova_table)
```

```python
# Perform Tukey's HSD post-hoc test for pairwise comparisons of age based on combined sex and who groups
from statsmodels.stats.multicomp import pairwise_tukeyhsd

tukey = pairwise_tukeyhsd(df['age'], df['sex'] + df['who'], alpha=0.05)

tukey.summary()
```

```python
# Create a bar plot to visualize age by 'who' and 'sex'
sns.barplot(df, x = 'who', y ='age', hue='sex')
```

```python
# Generate synthetic data for illustrating different types of correlation
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)

x = np.random.rand(100)
y_pos = x + np.random.normal(0, 0.1, 100)

x_neutral = np.random.rand(100)
y_neutral = np.random.normal(0.5, 0.1, 100)

y_neg = -x + np.random.normal(0, 0.1, 100)
```

```python
# Visualize positive, neutral, and negative correlations using scatter plots
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.scatter(x, y_pos, alpha=0.7)
plt.title('Positive Correlation')
plt.xlabel('X')
plt.ylabel('Y')

plt.subplot(1, 3, 2)
plt.scatter(x_neutral, y_neutral, alpha=0.7, color='green')
plt.title('Neutral Correlation')
plt.xlabel('X')
plt.ylabel('Y')

plt.subplot(1, 3, 3)
plt.scatter(x, y_neg, alpha=0.7, color='red')
plt.title('Negative Correlation')
plt.xlabel('X')
plt.ylabel('Y')

plt.tight_layout()
plt.show()
```
