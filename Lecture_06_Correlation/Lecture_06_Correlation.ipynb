# **Correlation**

**example of Pearson's correlation coefficient**

```python
import pandas as pd
import numpy as np
def pearson(x, y):
  x_mean = np.mean(x)
  y_mean = np.mean(y)
  x_std = np.std(x)
  y_std = np.std(y)
  n = len(x)
  return sum((x-x_mean)*(y-y_mean))/(n*x_std*y_std)
```
