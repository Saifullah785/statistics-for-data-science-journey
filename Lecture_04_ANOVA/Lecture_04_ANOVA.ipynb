# **1. ONE-WAY ANOVA**

```python
import scipy.stats as stats

fertilizer1 = [20, 22, 19, 24, 25]
fertilizer2 = [28, 30, 27, 26, 29]
fertilizer3 = [18, 20, 22, 19, 24]

f_stat, p_value = stats.f_oneway(fertilizer1, fertilizer2, fertilizer3)

print("F-statistic:", f_stat)
print("p-value:", p_value)

if p_value < 0.05:
    print("Reject null hypothesis: The mean are not equal, as the p-value is less than 0.05.")
else:
    print("Accept null hypothesis: The means are equal, as the p-value: {p_val} is greater than 0.05")
```
