# Data Distribution

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
df = sns.load_dataset('tips')
df.head()
```

```python
plt.figure(figsize=(8, 6))
sns.distplot(df['total_bill'], kde=True)
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(data=df, x='size', kde=True)
```
